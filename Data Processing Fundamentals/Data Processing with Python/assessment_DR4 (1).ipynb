## Exercício 1

### Contexto
Uma transportadora identifica cada encomenda por um código de rastreio de 8 caracteres no formato `OONNNNUF`. Os 2 primeiros caracteres indicam a **origem** (`BR` para nacional, `IN` para internacional), os 4 seguintes indicam o **número** da encomenda e os 2 últimos indicam a **UF de destino**.

### Tarefa
Percorra os códigos recebidos, separe origem, número e destino usando fatiamento de strings e exiba uma linha de classificação para cada encomenda. Ao final, informe quantas encomendas são de origem nacional e quantas têm destino `SP`.

### Entrada
A entrada será uma string contendo códigos de rastreio separados por espaço. Cada código segue o formato `OONNNNUF`.

Exemplo de entrada:

```python
codigos = "BR1023SP BR0987RJ IN4456SP BR2201MG IN0012RJ BR3345SP"
```

### Saída
Uma linha por encomenda no formato `BR1023SP -> origem: BR | numero: 1023 | destino: SP`, seguida de duas linhas com os totais.

Exemplo de saída parcial:

```text
BR1023SP -> origem: BR | numero: 1023 | destino: SP
...
Encomendas de origem nacional: 4
Encomendas com destino SP: 3
```


````
# Escreva sua solução aqui
codigos = "BR1023SP BR0987RJ IN4456SP BR2201MG IN0012RJ BR3345SP"

lista_codigos = codigos.split()

total_nacional = 0
total_destino_sp = 0

for codigo in lista_codigos:
    origem = codigo[:2]
    numero = codigo[2:6]
    destino = codigo[6:]

    print(f"{codigo} -> origem: {origem} | numero: {numero} | destino: {destino}")

    if origem == "BR":
        total_nacional += 1
    if destino == "SP":
        total_destino_sp += 1

print(f"Encomendas de origem nacional: {total_nacional}")
print(f"Encomendas com destino SP: {total_destino_sp}")
````

### Tarefa
Escreva uma **função** que receba uma linha bruta e devolva uma tupla `(nome, email, telefone)` limpa: sem espaços nas pontas de cada campo e com o telefone contendo apenas dígitos. Percorra todas as linhas, aplique a função e gere um relatório formatado com f-strings.

### Entrada
A entrada será uma string multilinha. Cada linha segue o formato geral `nome ; email ; telefone`, podendo conter espaços extras e pontuação no telefone.

Exemplo de entrada:

```python
cadastro = """  Ana Souza ;ana@mail.com ; (11) 99999-8888
Bruno Lima;  bruno@mail.com;(21) 98888-7777  
 Carla Reis ; carla@mail.com ; (31) 3333-2211 """
```

### Saída
Uma linha por cadastro, alinhada, no formato `Ana Souza | ana@mail.com | 11999998888`.


````
# Escreva sua solução aqui
cadastro = """  Ana Souza ;ana@mail.com ; (11) 99999-8888
Bruno Lima;  bruno@mail.com;(21) 98888-7777  
 Carla Reis ; carla@mail.com ; (31) 3333-2211 """

def limpar_cadastro(linha):
    partes = linha.split(";")
    nome = partes[0].strip()
    email = partes[1].strip()
    telefone = "".join(caractere for caractere in partes[2] if caractere.isdigit())
    return (nome, email, telefone)

for linha in cadastro.strip().split("\n"):
    nome, email, telefone = limpar_cadastro(linha)
    print(f"{nome} | {email} | {telefone}")
````

## Exercício 6

### Contexto

Uma maratona de programação recebeu inscrições em um arquivo de texto. Cada linha tem dados de uma pessoa inscrita. A organização só aceita maiores de idade.

Como o arquivo foi preenchido manualmente, algumas linhas podem estar com erro: idade escrita como texto, campo faltando ou formato diferente do esperado. O programa deve processar as linhas válidas e ignorar as inválidas sem parar a execução.

### Tarefa

Escreva uma **função** que receba o nome do arquivo e devolva apenas os inscritos com idade maior ou igual a 18.

Leia o arquivo linha a linha. Para cada linha:

1. separe os campos `nome`, `idade` e `categoria`;
2. tente converter a idade para número inteiro;
3. descarte os menores de idade;
4. ignore linhas inválidas usando `try/except`;
5. grave os aprovados em `aprovados.txt` no formato `nome,categoria`;
6. exiba uma estatística ao final.

A estatística deve mostrar:

- total de linhas lidas;
- total de aprovados;
- total de linhas inválidas ignoradas.

### Entrada

A entrada será um arquivo `.txt`. Cada linha válida do arquivo terá o formato:

```text
nome,idade,categoria
```

Exemplo de conteúdo do arquivo:

```text
Ana,22,backend
Bruno,17,frontend
Carla,30,backend
Davi,dezesseis,mobile
Edu,25
Fabi,19,backend
```

No exemplo acima:

- `Davi,dezesseis,mobile` é inválida porque a idade não pode ser convertida para inteiro;
- `Edu,25` é inválida porque falta a categoria.

A célula abaixo cria um arquivo de exemplo chamado `inscricoes.txt`.

### Saída

Criação do arquivo `aprovados.txt` e impressão de três linhas:

```text
Total de linhas lidas: 6
Total de aprovados: 3
Total de linhas inválidas: 2
```

O arquivo `aprovados.txt` deve conter apenas os inscritos válidos e maiores de idade, no formato:

```text
Ana,backend
Carla,backend
Fabi,backend
```

````
# Célula de preparação — cria um arquivo de exemplo para teste

conteudo = """Ana,22,backend
Bruno,17,frontend
Carla,30,backend
Davi,16,mobile
Edu,25,frontend
Fabi,19,backend"""

with open("inscricoes.txt", "w", encoding="utf-8") as f:
    f.write(conteudo)

print("inscricoes.txt criado.")
````

````
def processar_inscricoes(nome_arquivo):
    total_lidas = 0
    total_aprovados = 0
    total_invalidas = 0
    aprovados = []

    with open(nome_arquivo, 'r', encoding='utf-8') as arquivo:
        for linha in arquivo:
            linha = linha.strip()
            if not linha:
                continue
            
            total_lidas += 1
            
            try:
                campos = linha.split(',')
                if len(campos) != 3:
                    raise ValueError
                
                nome = campos[0]
                idade = int(campos[1])
                categoria = campos[2]
                
                if idade >= 18:
                    aprovados.append(f"{nome},{categoria}\n")
                    total_aprovados += 1
            
            except ValueError:
                total_invalidas += 1

    with open("aprovados.txt", "w", encoding="utf-8") as f:
        f.writelines(aprovados)

    print(f"Total de linhas lidas: {total_lidas}")
    print(f"Total de aprovados: {total_aprovados}")
    print(f"Total de linhas inválidas: {total_invalidas}")

processar_inscricoes("inscricoes.txt")
````

## Exercício 7

### Contexto
A coordenação tem as notas de uma turma organizadas como uma matriz. Cada linha da matriz contém o nome do aluno e suas notas.

### Tarefa
Escreva uma **função** que calcule a média de um aluno. Percorra a matriz e produza um relatório com a média de cada aluno, a média geral da turma e o nome do aluno com a maior média. Ao final, em **um comentário no código**, justifique se uma estrutura intermediária para guardar `nome -> média` faria mais sentido como **lista** ou como **dicionário**, e por quê.

### Entrada
A entrada será uma lista de listas. Cada lista interna segue o formato `[nome, nota1, nota2, nota3]`.

Exemplo de entrada:

```python
turma = [
    ["Ana", 7.0, 8.5, 6.0],
    ["Bruno", 9.0, 9.5, 8.0],
    ["Carla", 5.0, 6.0, 7.5],
    ["Davi", 8.0, 7.0, 9.0],
]
```

### Saída
Um relatório formatado com a média de cada aluno com uma casa decimal, a média geral da turma e o destaque de maior média. O comentário de justificativa faz parte da avaliação.


````
turma = [
    ["Ana", 7.0, 8.5, 6.0],
    ["Bruno", 9.0, 9.5, 8.0],
    ["Carla", 5.0, 6.0, 7.5],
    ["Davi", 8.0, 7.0, 9.0],
]

def calcular_media(notas):
    return sum(notas) / len(notas)

soma_geral = 0
maior_media = -1
destaque = ""
relatorio = []

for aluno_dados in turma:
    nome = aluno_dados[0]
    notas = aluno_dados[1:]
    media = calcular_media(notas)
    
    soma_geral += media
    relatorio.append(f"Aluno(a): {nome} | Média: {media:.1f}")
    
    if media > maior_media:
        maior_media = media
        destaque = nome

media_geral = soma_geral / len(turma)

for linha in relatorio:
    print(linha)

print(f"\nMédia geral da turma: {media_geral:.1f}")
print(f"Destaque: {destaque} com média {maior_media:.1f}")
````

### Para armazenar a relação "nome -> média", um dicionário (ex: {"Ana": 7.2, "Bruno": 8.8}) faria muito mais sentido do que uma lista. 

### Dicionários são estruturas de dados baseadas em chaves e valores, o que é semântica e tecnicamente ideal para mapear um identificador único (o nome do aluno) ao seu respectivo valor (a média).

## Exercício 8

### Contexto
Um e-commerce exporta os pedidos do dia em JSON. Cada pedido tem o cliente e uma lista de itens comprados. Cada item tem produto, quantidade e preço unitário.

### Tarefa
Escreva uma **função** que calcule o valor total de um pedido, somando `qtd * preco` de cada item. Leia o arquivo JSON, calcule o total de cada pedido, mantenha apenas os pedidos com total acima de R$ 100,00 e produza o valor total gasto por cliente considerando apenas esses pedidos.

### Entrada
A entrada será um arquivo `.json` contendo uma lista de pedidos. Cada pedido terá o formato geral:

```json
{"cliente": "Ana", "itens": [{"produto": "Fone", "qtd": 2, "preco": 80.0}]}
```

A célula abaixo cria um arquivo de exemplo chamado `pedidos.json`.

### Saída
A lista dos pedidos acima de R$ 100,00, com cliente e total, e o total acumulado por cliente.


````
# Célula de preparação — cria um arquivo JSON de exemplo para teste

import json

pedidos = [
    {"cliente": "Ana", "itens": [{"produto": "Fone", "qtd": 2, "preco": 80.0},
                                  {"produto": "Mouse", "qtd": 1, "preco": 45.0}]},
    {"cliente": "Bruno", "itens": [{"produto": "Monitor", "qtd": 1, "preco": 150.0}]},
    {"cliente": "Ana", "itens": [{"produto": "Cadeira", "qtd": 1, "preco": 220.0}]},
    {"cliente": "Carla", "itens": [{"produto": "Teclado", "qtd": 3, "preco": 120.0}]},
]

with open("pedidos.json", "w", encoding="utf-8") as f:
    json.dump(pedidos, f)

print("pedidos.json criado.")
````

````
# Escreva sua solução aqui
import json

def calcular_total_pedido(pedido):
    return sum(item["qtd"] * item["preco"] for item in pedido["itens"])

with open("pedidos.json", "r", encoding="utf-8") as f:
    dados_pedidos = json.load(f)

pedidos_filtrados = []
totais_por_cliente = {}

for pedido in dados_pedidos:
    total = calcular_total_pedido(pedido)
    if total > 100:
        pedidos_filtrados.append({"cliente": pedido["cliente"], "total": total})
        
        cliente = pedido["cliente"]
        if cliente in totais_por_cliente:
            totais_por_cliente[cliente] += total
        else:
            totais_por_cliente[cliente] = total

print("Pedidos acima de R$ 100,00:\n")
for p in pedidos_filtrados:
    print(f"Cliente: {p['cliente']} | Total: R$ {p['total']:.2f}")

print("\nTotal gasto por cliente:")
for cliente, total in totais_por_cliente.items():
    print(f"{cliente}: R$ {total:.2f}")
````

## Exercício 10

### Contexto
Um e-commerce armazena suas vendas em um arquivo JSON, com **uma linha por venda**: cada venda traz o cliente, o produto, a quantidade e o mês. Esse formato é bom para registrar, mas ruim para análise. A equipe quer uma visão **pivotada**: total vendido de cada produto, mês a mês.

### Tarefa
Escreva uma **função** que receba a lista de vendas e devolva um dicionário pivotado no formato `{mes: {produto: total_qtd}}`. Em seguida, para cada mês, identifique também o **produto mais vendido daquele mês**.

Leia o arquivo `vendas.json` usando `try/except`, exibindo uma mensagem clara caso o arquivo não exista ou não possa ser lido.

Imprima o resultado em **ordem cronológica de mês**, mostrando o dicionário do mês e o produto mais vendido logo em seguida.

### Entrada
A entrada será um arquivo `.json` contendo uma lista de vendas. Cada venda tem o formato geral:

```json
{"cliente": "Ana", "produto": "Fone", "qtd": 2, "mes": "2026-04"}
```

A célula abaixo cria um arquivo de exemplo chamado `vendas.json`.

### Saída
Para cada mês, em ordem cronológica, uma linha mostrando o dicionário pivotado daquele mês e qual foi o produto mais vendido.

Exemplo de saída esperada para os dados de teste:

```text
2026-04 -> {'Fone': 2, 'Monitor': 1, 'Teclado': 3} | mais vendido: Teclado (3)
2026-05 -> {'Fone': 4, 'Mouse': 2} | mais vendido: Fone (4)
```

````
# Célula de preparação — cria um arquivo JSON de exemplo para teste

import json

vendas = [
    {"cliente": "Ana",   "produto": "Fone",    "qtd": 2, "mes": "2026-04"},
    {"cliente": "Bruno", "produto": "Monitor", "qtd": 1, "mes": "2026-04"},
    {"cliente": "Ana",   "produto": "Fone",    "qtd": 1, "mes": "2026-05"},
    {"cliente": "Carla", "produto": "Teclado", "qtd": 3, "mes": "2026-04"},
    {"cliente": "Ana",   "produto": "Mouse",   "qtd": 2, "mes": "2026-05"},
    {"cliente": "Bruno", "produto": "Fone",    "qtd": 1, "mes": "2026-05"},
    {"cliente": "Carla", "produto": "Fone",    "qtd": 2, "mes": "2026-05"},
]

with open("vendas.json", "w", encoding="utf-8") as f:
    json.dump(vendas, f)

print("vendas.json criado.")
````

````
# Escreva sua solução aqui
import json

def pivotar_vendas(lista_vendas):
    vendas_pivot = {}
    for venda in lista_vendas:
        mes = venda["mes"]
        produto = venda["produto"]
        qtd = venda["qtd"]
        
        if mes not in vendas_pivot:
            vendas_pivot[mes] = {}
            
        if produto not in vendas_pivot[mes]:
            vendas_pivot[mes][produto] = 0
            
        vendas_pivot[mes][produto] += qtd
        
    return vendas_pivot

try:
    with open("vendas.json", "r", encoding="utf-8") as arquivo:
        dados_vendas = json.load(arquivo)
        
    vendas_agrupadas = pivotar_vendas(dados_vendas)
    
    for mes in sorted(vendas_agrupadas.keys()):
        produtos_do_mes = vendas_agrupadas[mes]
        
        produto_top = max(produtos_do_mes, key=produtos_do_mes.get)
        qtd_top = produtos_do_mes[produto_top]
        
        print(f"{mes} -> {produtos_do_mes} | mais vendido: {produto_top} ({qtd_top})")

except FileNotFoundError:
    print("Erro: O arquivo 'vendas.json' não foi encontrado. Verifique se ele foi criado corretamente.")
except json.JSONDecodeError:
    print("Erro: O arquivo 'vendas.json' está corrompido ou não possui um formato JSON válido.")
except Exception as e:
    print(f"Ocorreu um erro inesperado ao ler o arquivo: {e}")
````

Isso indica que o usuário tentou de novo e falhou de novo, sem recuperar — um sinal de problema persistente.

A ordem dos eventos no arquivo importa. Você precisa percorrer os eventos na ordem em que aconteceram, mantendo o **último nível registrado para cada usuário**, e marcar um incidente sempre que aparecer um `ERROR` para um usuário cujo último evento também foi `ERROR`.

### Tarefa
Escreva uma **função** que receba o nome do arquivo de log e devolva uma lista de incidentes. Cada incidente é um dicionário com `usuario`, `timestamp` e `acao` do segundo `ERROR` consecutivo (o que fechou o incidente).

Imprima ao final a lista de incidentes detectados e o total. Atenção: se um usuário tem `ERROR`, depois um `INFO` ou `WARN`, e depois outro `ERROR`, isso **não é** um incidente, porque houve recuperação entre eles.

### Entrada
A entrada será um arquivo `.txt`. Cada linha do arquivo terá o formato `timestamp | nivel | usuario | acao`, onde `nivel` pode ser `INFO`, `WARN` ou `ERROR`. Os campos vêm separados por `|` e podem ter espaços extras nas pontas.

Exemplo de conteúdo do arquivo:

```text
2026-05-20 09:01 | INFO  | ana    | login
2026-05-20 09:05 | ERROR | bruno  | falha de pagamento
2026-05-20 09:20 | ERROR | bruno  | falha de pagamento

```

Considere os eventos sempre na ordem em que aparecem no arquivo.

A célula abaixo cria um arquivo de exemplo chamado `log.txt`.

### Saída
A lista de incidentes detectados (com `usuario`, `timestamp` e `acao` de cada um) e o total de incidentes.

Para o arquivo de exemplo, a saída esperada é:

```text
Incidentes detectados:
- bruno em 2026-05-20 09:20 (falha de pagamento)
- carla em 2026-05-20 09:55 (timeout na consulta)
Total: 2
```

````
# Célula de preparação — cria um arquivo de log de exemplo para teste

log = """2026-05-20 09:01 | INFO  | ana    | login
2026-05-20 09:05 | ERROR | bruno  | falha de pagamento
2026-05-20 09:12 | WARN  | ana    | sessao expirando
2026-05-20 09:15 | ERROR | carla  | timeout na consulta
2026-05-20 09:20 | ERROR | bruno  | falha de pagamento
2026-05-20 09:33 | INFO  | bruno  | retry
2026-05-20 09:40 | ERROR | ana    | senha invalida
2026-05-20 09:55 | ERROR | carla  | timeout na consulta
2026-05-20 10:02 | WARN  | carla  | rate limit
2026-05-20 10:15 | ERROR | bruno  | falha de pagamento"""

with open("log.txt", "w", encoding="utf-8") as f:
    f.write(log)

print("log.txt criado.")
````

````
# Escreva sua solução aqui
# Célula de preparação — cria um arquivo de log de exemplo para teste
def detectar_incidentes(arquivo_log):
    ultimo_evento_usuario = {}
    incidentes = []

    try:
        with open(arquivo_log, "r", encoding="utf-8") as arquivo:
            for linha in arquivo:
                linha = linha.strip()
                if not linha:
                    continue

                partes = [parte.strip() for parte in linha.split('|')]
                
                if len(partes) != 4:
                    continue
                
                timestamp = partes[0]
                nivel = partes[1]
                usuario = partes[2]
                acao = partes[3]

                if usuario in ultimo_evento_usuario:
                    if ultimo_evento_usuario[usuario] == "ERROR" and nivel == "ERROR":
                        incidente = {
                            "usuario": usuario,
                            "timestamp": timestamp,
                            "acao": acao
                        }
                        incidentes.append(incidente)
                
                ultimo_evento_usuario[usuario] = nivel

        return incidentes

    except FileNotFoundError:
        print(f"Erro: Arquivo '{arquivo_log}' não encontrado.")
        return []

incidentes_detectados = detectar_incidentes("log.txt")

print("Incidentes detectados:")
for incidente in incidentes_detectados:
    print(f"- {incidente['usuario']} em {incidente['timestamp']} ({incidente['acao']})")

print(f"Total: {len(incidentes_detectados)}")
````

### Entrada
A entrada será uma lista de dicionários. Cada dicionário tem as chaves `nome` (string), `email` (string, possivelmente com case diferente ou espaços nas pontas) e `telefone` (string, possivelmente com parênteses, espaços ou traços).

Exemplo de entrada:

```python
cadastros = [
    {"nome": "Ana Silva",   "email": "ana@mail.com",     "telefone": "11999998888"},

    {"nome": "ANA",         "email": "ANA@MAIL.COM ",    "telefone": "11000001111"},
    {"nome": "Carla Reis",  "email": "carla@mail.com",   "telefone": "3133332211"},
    {"nome": "Bruno B.",    "email": "outro@mail.com",   "telefone": "21988887777"},
]
```

### Saída
Três informações: a contagem de cadastros únicos preservados, a contagem de descartados, e a lista dos descartados com nome e motivo (`"email"` ou `"telefone"`).


````
cadastros = [
    {"nome": "Ana Silva",   "email": "ana@mail.com",     "telefone": "11999998888"},
    {"nome": "Bruno Lima",  "email": "bruno@mail.com",   "telefone": "(21) 98888-7777"},
    {"nome": "ANA",         "email": "ANA@MAIL.COM ",    "telefone": "11000001111"},
    {"nome": "Carla Reis",  "email": "carla@mail.com",   "telefone": "3133332211"},
    {"nome": "Bruno B.",    "email": "outro@mail.com",   "telefone": "21988887777"},
]

# Escreva sua solução aqui
def remover_duplicatas(lista_cadastros):
    emails_vistos = set()
    telefones_vistos = set()
    cadastros_unicos = []
    descartados = []

    for cadastro in lista_cadastros:
        email_normalizado = cadastro["email"].strip().lower()
        telefone_normalizado = "".join(c for c in cadastro["telefone"] if c.isdigit())
        
        if email_normalizado in emails_vistos:
            descartados.append({"nome": cadastro["nome"], "motivo": "email"})
        elif telefone_normalizado in telefones_vistos:
            descartados.append({"nome": cadastro["nome"], "motivo": "telefone"})
        else:
            emails_vistos.add(email_normalizado)
            telefones_vistos.add(telefone_normalizado)
            cadastros_unicos.append(cadastro)
            
    return cadastros_unicos, descartados

unicos, descartados = remover_duplicatas(cadastros)

print(f"Cadastros únicos preservados: {len(unicos)}")
print(f"Cadastros descartados: {len(descartados)}")
print("\nLista de descartados:")
for item in descartados:
    print(f"- {item['nome']} (motivo: {item['motivo']})")
````

````
import requests
import json

# 1. Cole aqui sua chave de API do OpenRouter
chave_api = ""

url = "https://openrouter.ai/api/v1/chat/completions"

pergunta = "Explique em uma frase o que é uma API."


# 2. Monte os cabeçalhos da requisição (use Authorization: Bearer + chave_api)
headers = {
    "Authorization": f"Bearer {chave_api}",
    "Content-Type": "application/json"
}


# 3. Monte o corpo da requisição com o modelo "openrouter/free" e a pergunta
payload = {
    "model": "openrouter/free",
    "messages": [
        {
            "role": "user",
            "content": pergunta
        }
    ]
}


# 4. Faça a requisição POST para a URL acima, enviando headers e corpo,
#    e guarde o retorno em uma variável chamada "resposta"
resposta = requests.post(url, headers=headers, data=json.dumps(payload))


# 5. Forma de acessar o texto respondido pela IA:
dados = resposta.json()
resposta_ia = dados["choices"][0]["message"]["content"]
print(dados)
print(resposta_ia)
````
