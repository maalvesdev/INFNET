```
import requests
#usar a url estava dando erro 503 por conta de problemas do deepnote
#entao coloquei o endereco direto em vez da url
base_url = "http://127.0.0.1:8080"
endpoint = f"{base_url}/produtos-comuns"

try:
    response = requests.get(endpoint)
    
    if response.status_code == 200:
        produtos_comuns = response.json()
        print("Produtos em comum entre as duas filiais:")
        for produto in produtos_comuns:
            print(f"- {produto}")
    else:
        print(f"Falha na requisição. Código de erro: {response.status_code}")

except requests.exceptions.ConnectionError:
    print("Erro de conexão: Certifique-se de que a célula do Exercício 14 (o servidor) está rodando no outro notebook!")
```
